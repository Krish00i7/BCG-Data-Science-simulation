```python
plt.figure(figsize=(6,4))

sns.countplot(x="churn", data=client)

plt.title("Churn Distribution")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Count")

plt.show()
```

```python
key_numeric_cols = [
    "cons_12m", "cons_gas_12m", "cons_last_month",
    "forecast_cons_12m", "forecast_cons_year",
    "forecast_meter_rent_12m", "imp_cons",
    "margin_gross_pow_ele", "margin_net_pow_ele",
    "net_margin", "num_years_antig", "pow_max"
]

skewness = client[key_numeric_cols].skew()

print(skewness)

plt.figure(figsize=(10,5))
skewness.plot(kind="bar")
plt.title("Skewness of Numeric Features")
plt.ylabel("Skewness")
plt.xticks(rotation=45)
plt.show()
```

```python
for col in ["channel_sales", "origin_up", "has_gas"]:
    if col in client.columns:
        plt.figure(figsize=(6,4))

        client[col].value_counts().plot(kind="bar")

        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)

        plt.show()
```

```python
price_numeric_cols = price.select_dtypes(include=np.number).columns

price[price_numeric_cols].hist(figsize=(12,8))

plt.tight_layout()
plt.show()
```

```python
corr_matrix = client.select_dtypes(include=np.number).corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
client["tenure_years"] = (
    (client["date_activ"].max() - client["date_activ"]).dt.days
    / 365.25
)

plt.figure(figsize=(6,4))

plt.hist(client["tenure_years"], bins=30)

plt.title("Customer Tenure Distribution")
plt.xlabel("Tenure (Years)")
plt.ylabel("Count")

plt.show()
```

```python
sns.boxplot(x="churn", y="tenure_years", data=client)

plt.title("Customer Tenure vs Churn")
plt.show()
```
